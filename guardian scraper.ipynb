```python
import requests
import os
from bs4 import BeautifulSoup
```

```python
url='https://www.theguardian.com/uk'
req=requests.get(url, verify=False)
content=req.text
soup=BeautifulSoup(content)
```

```python
soup
```

```python
import pandas as pd
headlines = soup.find_all("a", class_="u-faux-block-link__overlay js-headline-text")

headline_texts = [headline.text for headline in headlines]
# Create a dataframe with the headline text
df = pd.DataFrame({'Headline': headline_texts})
# Print the dataframe
#print(df)
```

```python
# Import the wordcloud library
from wordcloud import WordCloud
# Join the different processed titles together.
long_string = ','.join(list(df['Headline'].values))
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue', scale=2)
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
wordcloud.to_image()
```

BBC

```python
url='https://www.bbc.co.uk/news'
req=requests.get(url, verify=False)
content=req.text
soup=BeautifulSoup(content)
```

```python
import pandas as pd
#headlines = soup.find_all("a", class_="gs-c-promo-heading gs-o-faux-block-link__overlay-link gel-pica-bold nw-o-link-split__anchor")
headlines = soup.find_all("a", class_="gs-c-promo-heading gs-o-faux-block-link__overlay-link gel-pica-bold nw-o-link-split__anchor")
headline_texts = [headline.text for headline in headlines]

# Create a dataframe with the headline text
df = pd.DataFrame({'Headline': headline_texts})

# Print the dataframe
print(df)
```

```python
# Import the wordcloud library
from wordcloud import WordCloud, STOPWORDS
# Join the different processed titles together.
long_string = ','.join(list(df['Headline'].values))
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue', scale=2, stopwords=STOPWORDS, collocations=False)
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
wordcloud.to_image()
```
